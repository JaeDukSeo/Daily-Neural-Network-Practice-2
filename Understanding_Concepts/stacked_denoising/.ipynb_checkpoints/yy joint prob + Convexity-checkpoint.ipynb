```python
# import lib
import numpy   as np
import seaborn as sns; sns.set(style="white", color_codes=True)
import matplotlib.pyplot as plt
```

```python
# draw some joint distribution
one   = np.random.randn(9000)
two   = np.random.uniform(-3,3,9000)
three = np.random.binomial(10,0.5,9000)

plt.figure(figsize=(20,10))
sns.jointplot(x=one,y=two, kind="kde", space=0, color="g") 
plt.figure(figsize=(20,10))
sns.jointplot(x=one,y=three, kind="kde", space=0, color="g") 
plt.show()
```

```python
# experiment
def gaussian(x, mu=0, sig=1.0):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

random_variable_x = np.asarray(range(-10,10))
random_variable_y = np.asarray(range(-15,5))
random_variable_z = np.asarray(range(-5,15))

rand_x_gau = np.asarray([gaussian(x) for x in random_variable_x])
rand_y_gau = np.asarray([gaussian(x) for x in random_variable_y])
rand_z_gau = np.asarray([gaussian(x) for x in random_variable_z])

temp = np.concatenate((np.array(range(-15,-10)),random_variable_x,np.array(range(11,15))))
print(temp)

print(random_variable_y)
print(random_variable_z)
```
