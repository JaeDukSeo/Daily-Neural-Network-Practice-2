```python
import numpy as np
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
import pandas as pd
np.random.seed(678)
sns.set_style("whitegrid")
%matplotlib inline
```

```python
data = load_breast_cancer()
df_data = pd.DataFrame(data.data,columns=data.feature_names)
df_data['Target'] = data.target
plt.figure(figsize=(20, 20))
sns.heatmap(np.around(df_data.corr(),2),  annot=True, vmax=.3, center=0,square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
df_data.columns
```

```python
from matplotlib import cm
df_features = df_data[['worst concave points','worst perimeter']]
df_targets = df_data[['Target']]
color_dict ={0:'blue',1:'red'}
plt.figure(figsize=(15, 15))
df_features.plot.scatter(x='worst concave points',y='worst perimeter',
                         c=df_targets['Target'].apply(lambda x: color_dict[x]),ax = plt.gca())
plt.show()
```

```python
classification_vector_1 = np.asarray([0.0,100])
classification_vector_1_mag = np.sqrt(classification_vector_1.dot(classification_vector_1))
classification_vector_2 = np.asarray([0.1,0])
classification_vector_2_mag = np.sqrt(classification_vector_2.dot(classification_vector_2))
classification_matrix = np.asarray([classification_vector_1,classification_vector_2])
classification_matrix_mag = np.asarray([classification_vector_1_mag,classification_vector_2_mag])

df_features_np = df_features.values
df_targets_np = df_data[['Target']].values
df_features_added = np.vstack((df_features_np,classification_vector_1))
df_features_added = np.vstack((df_features_added,classification_vector_2))
df_targets_np = np.vstack((df_targets_np,[3]))
df_targets_np = np.vstack((df_targets_np,[3]))
cmap = sns.cubehelix_palette(as_cmap=True)
plt.figure(figsize=(15, 15))
plt.scatter(df_features_added[:,0],df_features_added[:,1],c=np.squeeze(df_targets_np), cmap=cmap)
plt.show()
```

```python
resulted_array = []

for xx in df_features_np:
    temp = classification_matrix.dot(xx)/classification_matrix_mag
    
    if temp.sum() < 100:
        resulted_array.append(0)
    else:
        resulted_array.append(1)

resulted_array = np.expand_dims(np.asarray(resulted_array),1)
resulted_array = np.vstack((resulted_array,[3]))
resulted_array = np.vstack((resulted_array,[3]))
cmap = sns.cubehelix_palette(as_cmap=True)
plt.figure(figsize=(15, 15))
plt.title("classification two vectors")
plt.scatter(df_features_added[:,0],df_features_added[:,1],c=np.squeeze(resulted_array), cmap=cmap)
plt.show()
```
